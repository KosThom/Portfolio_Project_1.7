##### Goal: To scrape Top-3000 Drama Movies from IMDB, based on Revenue (also showing runtime, imdb rating, number of votes) and copy the data to an excel file

```python
from bs4 import BeautifulSoup
import requests, openpyxl, re
# Import Libraries that we are going to use

excel = openpyxl.Workbook()        
# Create a new excel file
print(excel.sheetnames)            
# Check how many sheets there are in the excel file we created
sheet = excel.active               
# Make sure we are working on the active sheet (where we are going to load the data)
sheet.title = 'Movies-TV-Shows'    
# Change the name of the active sheet
print(excel.sheetnames)            
# Check whether the sheet name has been changed
sheet.append(['Title', 'Year of Release', 'Runtime (minutes)', 'IMDB Rating (Out of 10)', 'Number of Votes', 'Revenue'])
# Create headers in the active excel sheet

try:
    headers = {"Accept-Language": "en-US,en;q=0.5"} 
    # In order to avoid returning the movie titles in Greek 
    our_url = 'https://www.imdb.com/search/title/?genres=drama&sort=boxoffice_gross_us,desc&start={}&explore=title_type,genres&ref_=adv_nxt'
    # This is the url from which we are going to get the data. The {} refers to a number that changes depending 
    # on the first movie shown in the page (1, 51, 101 etc). 50 Movies are shown in each page, so for the last page 
    # this number should be 2951, since we are interested in getting 3,000 movies
    
    for x in (1,51,101,151,201,251,301,351,401,451,501,551,601,651,701,751,801,851,901,951,1001,1051,1101,1151,
              1201,1251,1301,1351,1401,1451,1501,1551,1601,1651,1701,1751,1801,1851,1901,1951,2001,2051,2101,2151,
              2201,2251,2301,2351,2401,2451,2501,2551,2601,2651,2701,2751,2801,2851,2901,2951):
        scrape_url = our_url.format(x)
        source = requests.get(scrape_url, headers=headers)
        # We are using requests so we can extract the htlm code from the webpage                          
        source.raise_for_status()
        # In order to check if the url we defined is valid, there is the raise_for_status method
        
        soup = BeautifulSoup(source.text,'html.parser')
        # We are pulling in the entire html code of the webpage
        movies = soup.find('div', class_="lister-list").find_all('div', class_="lister-item-content")
        # We are defining the path in which the data we need are located for each movie
        
        for movie in movies:
            
        
            name = movie.find('h3', class_="lister-item-header").a.text
        
            year = movie.find('span', class_="lister-item-year text-muted unbold").text
        
            runtime = movie.find('span', class_="runtime").text.split(" ")[0]
        
            rating = movie.find('div', class_="inline-block ratings-imdb-rating").text.strip('()')
        
            votes = movie.find("p", class_="sort-num_votes-visible").text.split("|")[0].split(":")[1]

            revenue = movie.find("p", class_="sort-num_votes-visible").text.split("|")[1].split(":")[1]
        
            
            print(name, (int(re.findall('(\d+)',year)[0])), runtime, rating, votes, revenue) 
            # use of regular expression, in order to get only the numeric part of the year (excluding "()" and other text)
            
            sheet.append([name, int(re.findall('(\d+)',year)[0]), runtime, rating, votes, revenue]) 
            # load the values extracted from the website into the excel file
                
        
        
except Exception as e:
    print(e)
    
excel.save('3000_IMDB_Scraping_Project.xlsx') 
# save the excel file
```
